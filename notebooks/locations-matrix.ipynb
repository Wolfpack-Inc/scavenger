```python
# global imports
from scavenger import *
import numpy as np
import pandas as pd
import geopy.distance
```

### Load data from db to df

```python
locations = (Locations
             .select(Locations.street, Locations.dis_long, Locations.dis_lat)
             .dicts())

df = pd.DataFrame(locations, columns=['street', 'dis_long', 'dis_lat'])
df.head()
```

```python
df.shape
```

```python
df['coords'] = df.apply(lambda row: (row['dis_long'], row['dis_lat']), axis=1)
df.head()
```

### Initialize matrix

```python
matrix_distances = np.zeros((df.shape[0], df.shape[0]))
matrix_distances
```

```python
def get_distance(coords1, coords2):
    return geopy.distance.geodesic(coords1, coords2).km
```

### Looping over Dataframe and insert distances into matrix

```python
for index, row in df.iterrows():
    idx1 = index
    coords1 = row['coords']
    print(idx1)
    for index, row in df.iterrows():
        coords2 = row['coords']
        idx2 = index
        matrix_distances[idx1, idx2] = get_distance(coords1, coords2)
```

```python
matrix_distances
```

### Creating dataframe with streetnames as indices and columns from matrix

```python
df_distances = pd.DataFrame(matrix_distances, index=df['street'], columns=df['street'])
df_distances.head()
```
